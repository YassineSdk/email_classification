### **Exploratory Analysis: Key Features in Bot Detection**

- **`default_profile`**:  
  The bar plot shows that **50%** of accounts with the default profile are bots, while the other half are human. In contrast, among accounts with a personalized profile, only **20%** are bots and **80%** are humans.  
   This suggests that having a default profile is a **potential indicator** of bot-like behavior.

- **`default_profile_image`**:  
  Around **70%** of bots use the default Twitter profile image, whereas only about **25%** of accounts with a custom profile image are bots.  
   Bots are far more likely to retain the default image.

- **`geo_enabled`**:  
  Most human accounts have the **geo-location feature enabled**. Among accounts with this feature **disabled**, the distribution between bots and humans appears to be relatively **even**.  
   Enabling geo-location may be more common among human users.

- **`verified`**:  
  Over **90%** of verified accounts are **human**.  
   Verification is a **strong signal** of an authentic (human) account.

---

###  Hypothesis Testing

To confirm these observations, we apply the **Chi-squared test of independence** to determine whether these features (`default_profile`, `default_profile_image`, `geo_enabled`, `verified`) are **statistically associated** with `account_status` (bot or human).

- A **significant Chi-squared result (p < 0.05)** indicates that the variables are **not independent**, meaning the feature is **statistically relevant** in distinguishing bots from humans.




```python
from scipy.stats import chi2_contingency

for col in categ_col:
    data = pd.crosstab(df[col],df['account_type'])
    chi_2,p_value,_,_= chi2_contingency(data)
    D(M(f'#### account_status and {col}'))
    D(M(f'the contribution of ``{col}`` : **{chi_2:.2f}**'))
    if p_value < .05 :
        D(M(f'the association of the account status and the {col} is **statistically significant** this means that the two columns are dependants'))
    else :
        D(M(f'the association of the account status and the {col} is **not statistically significant** this means that the two columns are independants'))
    D(M('---'))    
```

**conclusion**
- *Based on the Chi-squared test of independence, we are confident in our hypothesis that the categorical features default_profile, default_profile_image, geo_enabled, and verified play a key role in distinguishing between bot and human accounts.*

*The results indicate that these variables are not independent from the account_status, suggesting they carry significant discriminative power in identifying account types.*
